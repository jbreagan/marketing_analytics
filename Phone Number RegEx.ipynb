```python
import re
import pandas as pd
import logging

df = pd.read_csv('renter_phone_number2.txt',delimiter="\t")
df.head()
```

```python
def format_numbers(dataframe):
    renter_phone_data = dataframe
    logging.info("Renter data imported w/ length " + str(len(dataframe)))
    renter_phone_data["phone_number"] = renter_phone_data["phone_number"].astype(str)
    renter_phone_data["phone_number"] = renter_phone_data["phone_number"].str.replace(
        r"[^0-9]+", ""
    )
    logging.info("Renter data without extra characters updated w/ length " + str(len(dataframe)))

    trimmed_number = renter_phone_data["phone_number"].str[:-1]
    last_number = renter_phone_data["phone_number"].str[-1]
    renter_phone_data["phone_number"] = (
        trimmed_number.str.replace("(\d)(?=(\d{3})+(?!\d))", r"\1-") + last_number
    ).astype(str)
    logging.info("Renter data formatted w/ length " + str(len(dataframe)))

    renter_phone_data["phone_number"] = renter_phone_data["phone_number"].apply(
        lambda x: x[-12:] if len(x) > 11 else x
    )
    logging.info("Renter phone numbers shortened w/ length " + str(len(dataframe)))

    renter_phone_data = renter_phone_data.drop_duplicates(
        ["renter_id", "phone_number"]
    ).set_index("renter_id")
    logging.info("Renter data deduped w/ length " + str(len(dataframe)))
    return renter_phone_data
```
